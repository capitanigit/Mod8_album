```python
import pandas as pd
data = pd.read_html('https://www.officialcharts.com/chart-news/the-best-selling-albums-of-all-time-on-the-official-uk-chart__15551/', header=0)
df=data[0]
```

AD 1. Zamień nagłówki kolumn na polskie odpowiedniki

```python
df.rename(columns = {'TITLE' : 'TYTUŁ', 'ARTIST' : 'ARTYSTA', 'YEAR' : 'ROK', 'HIGH POSN' : 'MAX POZ'}, inplace = True)
df.head()
```

AD 2. Ilu pojedynczych artystów znajduje się na liście?

```python
df['ARTYSTA'].nunique()
```

AD 3. Które zespoły pojawiają się najczęściej na liście?

```python
df['ARTYSTA'].value_counts().head()
```

AD 4. Zmień nagłówki kolumn, tak aby każdy z nich rozpoczynał się od wielkiej litery, a pozostałe były wprowadzone małymi literami.

```python
df.rename(str.capitalize, axis='columns', inplace=True)
df.head()
```

AD 5. Wyrzuć z tabeli kolumnę ‘Max Poz’

```python
df.drop('Max poz',axis=1,inplace=True)
df.head()
```

AD 6. W którym roku wyszło najwięcej albumów znajdujących się na liście?

```python
df['Rok'].value_counts().head()
```

AD 7. Ile albumów wydanych między 1960 a 1990 rokiem włącznie znajduje się na liście?

```python
df[(df['Rok']>=1960) & (df['Rok']<=1990)].count()
```

AD 8. W którym roku wydany został najmłodszy album na liście?

```python
df.sort_values(by='Rok', ascending=False).head()
```

AD 9. Przygotuj listę najwcześniej wydanych albumów każdego artysty, który znalazł się na liście.

```python
df.sort_values('Rok').groupby('Artysta')[['Tytuł', 'Rok']].first()
```

AD 10. Listę zapisz do pliku csv

```python
df.to_csv('album.csv', index=False)
```
